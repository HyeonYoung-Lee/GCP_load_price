```python
# define output dict
output = {}
for machine_type in machine_type_list:
    output[machine_type] = {}
    for region in region_list:
        output[machine_type][region] = {}
        output[machine_type][region]['ondemand'] = None
        output[machine_type][region]['preemptible'] = None
```

```python
# extract compute engine
compute_engine_pricelist = {}
for description, value in pricelist.items():
    if description.split('-')[0] == 'CP' and description.split('-')[1] == 'COMPUTEENGINE':
        compute_engine_pricelist[description] = value
```

```python
def put_price(machine_type, regions):
    for region, price in regions.items():
        if price == 0:
            price = None
        if region in region_list:
            # preemptible
            if 'PREEMPTIBLE' in description :
                output[machine_type][region]['preemptible'] = price
            else:
                output[machine_type][region]['ondemand'] = price
```

```python
def extract_machine_type(machine_str, description):
    if machine_str.upper() in description :
        cpu_num = description.split('-')[5]
        machine_type = machine_str + '-' + cpu_num
        return machine_type
    else:
        return None
```

```python
for description, regions in compute_engine_pricelist.items():
    # N1, E2
    machine_str_list = ['n1-standard', 'n1-highmem', 'n1-highcpu', 'e2-standard', 'e2-highmem', 'e2-highcpu']
    for machine_str in machine_str_list:
        machine_type = extract_machine_type(machine_str, description)
        if machine_type != None:
            put_price(machine_type, regions)
    
    # N1 : f1-micro, g1-small
    if 'CP-COMPUTEENGINE-VMIMAGE-F1-MICRO' in description :
        put_price('f1-micro', regions)
    if 'CP-COMPUTEENGINE-VMIMAGE-G1-SMALL' in description :
        put_price('g1-small', regions)

    # E2 : e2-micro, e2-small, e2-medium
    if 'CP-COMPUTEENGINE-VMIMAGE-E2-MICRO' in description :
        put_price('e2-micro', regions)
    if 'CP-COMPUTEENGINE-VMIMAGE-E2-SMALL' in description :
        put_price('e2-small', regions)
    if 'CP-COMPUTEENGINE-VMIMAGE-E2-MEDIUM' in description :
        put_price('e2-medium', regions)
        
```

```python
def calculate_price(cpu_data, ram_data, machine_series, price_type):
    for machine_type, feature in machine_series.items():
        cpu_quantity = feature['cpu']
        ram_quantity = feature['ram']
        for cpu_region, cpu_price in cpu_data.items():
            for ram_region, ram_price in ram_data.items():
                if cpu_region == ram_region and cpu_region in region_list:
                    output[machine_type][cpu_region][price_type] = cpu_quantity*cpu_price + ram_quantity*ram_price
```
